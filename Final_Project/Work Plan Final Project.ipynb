# Proyecto final
## Descripción
Al operador de telecomunicaciones Interconnect le gustaría poder pronosticar su tasa de cancelación de clientes. Si se descubre que un usuario o usuaria planea irse, se le ofrecerán códigos promocionales y opciones de planes especiales. El equipo de marketing de Interconnect ha recopilado algunos de los datos personales de sus clientes, incluyendo información sobre sus planes y contratos.

### Servicios de Interconnect

Interconnect proporciona principalmente dos tipos de servicios:

1. Comunicación por teléfono fijo. El teléfono se puede conectar a varias líneas de manera simultánea.
2. Internet. La red se puede configurar a través de una línea telefónica (DSL, *línea de abonado digital*) o a través de un cable de fibra óptica.

Algunos otros servicios que ofrece la empresa incluyen:

- Seguridad en Internet: software antivirus (*ProtecciónDeDispositivo*) y un bloqueador de sitios web maliciosos (*SeguridadEnLínea*).
- Una línea de soporte técnico (*SoporteTécnico*).
- Almacenamiento de archivos en la nube y backup de datos (*BackupOnline*).
- Streaming de TV (*StreamingTV*) y directorio de películas (*StreamingPelículas*)

La clientela puede elegir entre un pago mensual o firmar un contrato de 1 o 2 años. Puede utilizar varios métodos de pago y recibir una factura electrónica después de una transacción.

### Descripción de los datos

Los datos consisten en archivos obtenidos de diferentes fuentes:

- `contract.csv` — información del contrato;
- `personal.csv` — datos personales del cliente;
- `internet.csv` — información sobre los servicios de Internet;
- `phone.csv` — información sobre los servicios telefónicos.

En cada archivo, la columna `customerID` (ID de cliente) contiene un código único asignado a cada cliente. La información del contrato es válida a partir del 1 de febrero de 2020.

## Preguntas y suposiciones
1. Me imagino que del paquete general donde se ofrecen los servicios de teléfono fijo e internet, muchos usuarios al no usar el teléfono fijo considerarían que no vale la pena pagar por el servicio y por ende darse de baja. ¿Cuántos de ellos ocupan verdaderamente ambos servicios en comparación con los que utilizan solo el internet?
2. Sería esencial comparar cuantos clientes contratan otros servicios además del básico y hacer un recuento de cuántas bajas existen si más del lado que solo adquieren dos servicios o más.
3. Puede que la razón por la que se den de baja los clientes sea porque no tienen acceso a un soporte técnico dentro del paquete básico y al tener un costo adicional prefieren pagar otro servicio con el que si haya esta opción.
4. ¿Influye con que streamings cuente el servicio para saber si el usuario decide o no quedarse?
5. ¿Influye el espacio de almacenamiento con el que cuenta el usuario para quedarse o no? ¿será muy poco espacio y requiere de más?
6. ¿De qué calidad es el sistema de seguridad de internet? ¿Se puede mejorar o promocionar junto con la instalación de intenet del pquete básico?
7. ¿Cómo influye el tiempo en el que una persona cuenta con un paquete de 1 o 2 años a los que solo rentan mensualmente? Comparar si hay mayor cantidad de usuarios que deciden irse más de una forma de o de la otra.
 

## Pasos a desarrollar
1. Contamos con cuatro archivos csv para resolver la tarea, asi que primero hay que checar en cada uno de ellos la presencia de valores nulos y duplicados asi como verificar que el tipo de dato de cada columna del dataframe sea el correcto. Después de  esto, hay que realizar un análisis exploratorio con ayuda de gráficos y diagramas de cajas para entender como se comportan las características con el fin de descubrir patrones y anomalías.

3. Después hay que definir para el modelo de machine learning las características y objetivo para después dividir en conjuntos de datos de entrenamiento y prueba en un formato de 70% y 30%. No obstante, antes de eso sería muy bueno checar si hay un desbalance de clases en el dataframe para saber si ocupar funciones de sobremuestreo o submuestreo ya que en un futuro puede afectar que una de las clases (como clientes que cancelan) tengan significativamente menos instancias que la otra (clientes que no cancelan).
4. El modelo es de clasificación supervisada, ya que el objetivo es predecir si un cliente cancelará o no cancelará su suscripción, es decir que se trata de un problema binario (cancelar/no cancelar). Por ello, lo mejor sería tratar con diferentes modelos (regresión logística, bosque aleatorio, XGBoost, LightGBM y-o redes neuronales) y verificar cual de ellos es el mas eficiente en velocidad y en el valor de sus métricas (exactitud, precision, recall, ROC-AUC).
5. En caso de que se necesitara se puede implementar validacion cruzada para  evaluar el rendimiento de los modelos y asi  evitar el sobreajuste.
